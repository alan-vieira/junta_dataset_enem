# Carregando e analisando os datasets

## ENEM 2009

```python
# a variável Q3 contém a informação cor/raça
colunas = [
            'NU_ANO',
            'TP_FAIXA_ETARIA',
            'TP_SEXO',
            'Q3',
            'TP_DEPENDENCIA_ADM_ESC',
            'NO_MUNICIPIO_ESC',
            'SG_UF_ESC',
            'NU_NOTA_CN',
            'NU_NOTA_CH',
            'NU_NOTA_LC',
            'NU_NOTA_MT',
            'NU_NOTA_REDACAO'
            ]

enem2009_df = pd.read_csv(
                            'raw/MICRODADOS_ENEM_2009.csv', 
                            sep=';', 
                            encoding='ISO-8859-1', 
                            usecols=colunas
                            )
```

```python
display(enem2009_df)
```

```python
# removendo as linhas dos caditados que não possuem notas das provas
enem2009_df.dropna(subset=[
                            'NU_NOTA_CN', 
                            'NU_NOTA_CH', 
                            'NU_NOTA_LC', 
                            'NU_NOTA_MT', 
                            'NU_NOTA_REDACAO'
                            ], inplace=True)
enem2009_df.head()
```

```python
# trocando o nome da coluna Q3 para TP_COR_RACA
enem2009_df.rename(columns = {'Q3': 'TP_COR_RACA'}, inplace = True)
enem2009_df.head()
```

```python
# padronizando a variável TP_COR_RACA para int, como nos outros datasets
enem2009_df['TP_COR_RACA'] = enem2009_df['TP_COR_RACA'].map({
                                                                'A':1,
                                                                'B':3, 
                                                                'C':2, 
                                                                'D':4, 
                                                                'E':5
                                                            })
```

```python
enem2009_df.head()
```

```python
# substituindo NaN por 0 na coluna TP_COR_RACA (o 0 representa a informação não declarada)
enem2009_df['TP_COR_RACA'] = enem2009_df['TP_COR_RACA'].fillna(0)
enem2009_df.head()
```

```python
enem2009_df['TP_COR_RACA'].value_counts()
```
